````python
import csv
import re

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')   
    
    
    firstname = [] #without x means that it is correct
    lastname = []
    x_birthmonth = [] # x signifies not yet cleaned
    birthday = []
    x_birthyear = []
    x_course = []
    x_phonenumber = []
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            firstname.append(row[0])
            lastname.append(row[1])
            x_birthmonth.append(row[2])
            birthday.append(row[3])
            x_birthyear.append(row[4])
            x_course.append(row[5])
            x_phonenumber.append(row[6])
        line_count += 1

# making all birth months "January" format (clean)
# made a dirty list then used [n for n in unique_names] and unique_names_dictionary = {name:None for name in unique_names}
# then manually inputted the correct format and made put the result in a dicitonary
# applicable for course as well
    unique_birthmonth_dictionary = {'1':'January','2':'February','3':'March','4':'April','5':'May','6':'June','7':'July',
                             '8':'August','9':'September','10':'October','11':'November','12':'December','January':'January',
                              'February':'February','March':'March','April':'April','May':'May','June':'June','July':'July',
                             'August':'August','September':'September','October':'October','November':'November',
                              'December':'December','Jan':'January','Feb':'February','Mar':'March','Apr':'April',
                              'May':'May','Jun':'June','Jul':'July','Aug':'August','Sep':'September',
                              'Oct':'October','Nov':'November','Dec':'December'}
    birthmonth = [re.sub(r''.join(month),r''.join(unique_birthmonth_dictionary[month]),month) for month in x_birthmonth]
    
# making all birth years "1990" format (clean)
    unique_birthyear_dictionary = {'97':'1997','98':'1998','99':'1999','00':'2000','01':'2001','1997':'1997','1998':'1998',
                             '1999':'1999','2000':'2000','2001':'2001'}
    birthyear = [re.sub(r''.join(year),r''.join(unique_birthyear_dictionary[year]),year) for year in x_birthyear]
    
# making all courses "BS Management" format (clean)
# similar style to birth month
    unique_course_dictionary = {'ME': 'BS Management Engineering','BS LM': 'BS Legal Management','MGT': 'BS Management',
                         'Communications Technology': 'BS Communications Technology Management','BS MGT': 'BS Management',
                         'Information Technology Entrepreneurship': 'BS Information Technology Entrepreneurship',
                         'BS CTM': 'BS Communications Technology Management','ITE': 'BS Information Technology Entrepreneurship',
                         'BS ITE': 'BS Information Technology Entrepreneurship','Management': 'BS Management',
                         'CTM': 'BS Communications Technology Management','LM': 'BS Legal Management','Legal Management': 'BS Legal Management',
                         'BS ME': 'BS Management Engineering','Management Engineering': 'BS Management Engineering'}
    course = [re.sub(r''.join(course),r''.join(unique_course_dictionary[course]),course) for course in x_course]
    
# making all phone numbers "+639xxxxxxxxx" format (clean)
# 2 steps
    almost_phonenumber = [re.sub(r'^(9[0-9][0-9])',r'+63 \1',number) for number in x_phonenumber]
    phonenumber = []
    for number in almost_phonenumber:
        phonenumber.append(number.replace(" ",""))
        

    clean = list(zip(firstname,lastname,birthmonth,birthday,birthyear,course,phonenumber))

# new csv file
with open('cleanrecweekdata.csv','w') as csv_file:
    csv_writer = csv.writer(csv_file,lineterminator='\n')
    csv_file.write('firstname,lastname,birthmonth,birthday,birthyear,course,phonenumber\n')
    csv_writer.writerows(clean)
````
